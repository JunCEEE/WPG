## Propagators for initial forward propagation for smoothing (L0=10, 20, 50m)

Here we have define distances, on which wavefront should be propagated fron 'zero' position.

**common_distances** - array of distances for searching waste position.

For smoothing wavefront we will froward-propagate initial wavefront to some distances 10,20,50 meters. For each distance we should specify forward- and back- propagation parameters. The best if this parameters will same for all distances. To propagate to 10 m. and 20 m. distances please uncomment the correspon

**multi_distance_parameters** - list of sets of propagation parameters.

```python
#Wavefront Propagation Parameters:
#[0]:  Auto-Resize (1) or not (0) Before propagation
#[1]:  Auto-Resize (1) or not (0) After propagation
#[2]:  Relative Precision for propagation with Auto-Resizing (1. is nominal)
#[3]:  Allow (1) or not (0) for semi-analytical treatment of quadratic phase terms at propagation
#[4]:  Do any Resizing on Fourier side, using FFT, (1) or not (0)
#[5]:  Horizontal Range modification factor at Resizing (1. means no modification)
#[6]:  Horizontal Resolution modification factor at Resizing
#[7]:  Vertical Range modification factor at Resizing
#[8]:  Vertical Resolution modification factor at Resizing
#[9]:  Type of wavefront Shift before Resizing (not yet implemented)
#[10]: New Horizontal wavefront Center position after Shift (not yet implemented)
#[11]: New Vertical wavefront Center position after Shift (not yet implemented)
    

common_distances = np.arange(-80,50,5.0)
common_distances = np.append(common_distances, np.arange(-20,10,2.0))
common_distances = np.append(common_distances, np.arange(-20,0,0.5))

common_distances = np.unique(common_distances)

multi_distance_parameters = []

# list_10 = common_distances - 10
# multi_distance_parameters.append({'forward_propagation_distance':10.0,
#     'forward_propagation_parameters': [ 0,  0,  1,  0,  0, 2.5, 1.0, 2.5, 1.0,  0,  0,   0],
#     'back_propagation_parameters': [ 0,  0,  1,  0,  0, 1.0, 1.0, 1.0, 1.0,  0,  0,   0],
#     'propagation_distances': list_10
#     })

# list_20 = common_distances - 20
# multi_distance_parameters.append({'forward_propagation_distance':20.0,
#     'forward_propagation_parameters': [ 0,  0,  1,  0,  0, 2.5, 1.0, 2.5, 1.0,  0,  0,   0],
#     'back_propagation_parameters': [ 0,  0,  1,  0,  0, 1.0, 1.0, 1.0, 1.0,  0,  0,   0],
#     'propagation_distances': list_20
#     })

list_50 = common_distances - 50
multi_distance_parameters.append({'forward_propagation_distance':50.0,
    'forward_propagation_parameters': [ 0,  0,  1,  0,  0, 2.5, 1.0, 2.5, 1.0,  0,  0,   0],
    'back_propagation_parameters': [ 0,  0,  1,  0,  0, 1.0, 1.0, 1.0, 1.0,  0,  0,   0],
    'propagation_distances': list_50
    })
```

Now we will start propagation process in multicore mode. This process will consume a lot of cpu and memory and disk resource.

For every forward-propagated distance will create separate directory (for example 10.0000 or 20.0000). In this directory will stored forward propagated wavefront (for example 10.0000.h5 or 50.0000.h5) and back-propagated files (for exmaple 50.0000\_-5.0000.h5, the froward propagated for 50 meters and then back propagated to 5 meters)

If any errors will rised please restart this cell. The calculated wavefronts will not be calculated, only missed wavefronts will calculated. You may add some additionall distances and recalculate. 

If yo change propagation parameters, you should delete all calculated wavefronts !!!

```python
# propagate wavefreonts with paramters from multi_distance_parameters
for propagation_parameters in multi_distance_parameters:
    print 'Forward propagation distance: {}'.format(propagation_parameters['forward_propagation_distance'])
    wf_file = forward_propagate(root_directory, propagation_parameters['forward_propagation_distance'],
                       propagation_parameters['forward_propagation_parameters'])
    
    arg_list = [(wf_file, distance, propagation_parameters['back_propagation_parameters']) 
                    for distance in propagation_parameters['propagation_distances']]
    files = [True,]
    while not all([x is None for x in files]):
        try:
            p = multiprocessing.Pool(processes = multiprocessing.cpu_count()-2) # reduce the processes value if you have problems with memory consumption.
            files = p.map(back_propagate, arg_list, chunksize=1)
            p.close()
        except MemoryError:
            gc.collect()
        finally:
            del p
            gc.collect()
```

## Analyse wavefront files to collect inforamtion abaout gaussina parameters

Now we will walk through directories, extract information from wavefronts and store it in 'collected_data.pkl' filen in each directory. Process may take a lot of time and consume cpu and memory. Reduce numper of cpu-cores if process fials. After data collected HDF5 files could be deleted.  

```python
#Lets find directory with numerical names with hdf5 files 
ls_data = os.listdir(root_directory)
dirs = []
for l in ls_data:
    full_path = os.path.join(root_directory,l)
    if os.path.isdir(full_path):
        try:
            x=np.float(l)
            dirs.append(full_path)
        except ValueError:
            pass
```

```python
#extract parameters from wavefront to python data files collected_data.pkl
def extract_parameters(file_name):
    def load_hdf5_tree(h5_group):
        """convert h5file to the tree"""
        out_dict = {}

        def add_item(name, obj):
            if not isinstance(obj, h5py.Group):
                out_dict.update({name.encode(): obj.value})

        h5_group.visititems(add_item)

        return out_dict

    try:
        
        res = {}
        
        with h5py.File(file_name, 'r') as h5:
            z_coord = h5['params']['Mesh']['zCoord'].value
            res['misc'] = load_hdf5_tree(h5['misc'])
            res['params']={'Mesh':{}}
            res['params']['Mesh']['xMin'] = h5['params']['Mesh']['xMin'].value
            res['params']['Mesh']['xMax'] = h5['params']['Mesh']['xMax'].value
            res['params']['Mesh']['nx'] = h5['params']['Mesh']['nx'].value

            res['params']['Mesh']['yMin'] = h5['params']['Mesh']['yMin'].value
            res['params']['Mesh']['yMax'] = h5['params']['Mesh']['yMax'].value
            res['params']['Mesh']['ny'] = h5['params']['Mesh']['ny'].value

            res['params']['Mesh']['sliceMin'] = h5['params']['Mesh']['sliceMin'].value
            res['params']['Mesh']['sliceMax'] = h5['params']['Mesh']['sliceMax'].value
            res['params']['Mesh']['nSlices'] = h5['params']['Mesh']['nSlices'].value

            res['params']['Mesh']['zCoord'] = h5['params']['Mesh']['zCoord'].value
            res['params']['photonEnergy'] = h5['params']['photonEnergy'].value
            res['misc']['gaussian_parameters/pulse_max_intensity'] = res['misc']['gaussian_parameters/pulse_intensity'].max()
            del res['misc']['gaussian_parameters/pulse_intensity']
    
        gc.collect()
        return {z_coord: res}
    except IOError:
        print 'File not found {}'.format(file_name)
        return None


for d in dirs:
#     if os.path.exists(os.path.join(d,'collected_data.pkl')):
#         print 'File {} exists. Skip directory'.format(os.path.join(d,'collected_data.pkl'))
#         continue
        
    files = glob.glob(os.path.join(d,'*.h5'))
    print 'Found {} files in {}'.format(len(files), d)

    p= multiprocessing.Pool(processes=multiprocessing.cpu_count()-2) # reduce the processes value if you have problems with memory consumption.
    res=p.map(extract_parameters, files, chunksize=1)
    p.close()
    del p
    
    data = {}
    for r in res:
        if not r is None:
            data.update(r)
    
    with open(os.path.join(d,'collected_data.pkl'),'w') as pf:
        pickle.dump(data,pf)
    
    del res 
    del data
    gc.collect()
```

## Plot summary.

Gaussian beam formula used for approximation:
$$ height*\exp \left[-\frac{1}{2}\left(\left(\frac{center\_x-x}{width\_x}\right)^2+\left(\frac{center\_y-y}{width\_y}\right)^2\right)\right]$$

** For 2D plots **: 

The *height* - shows the intensity of gaussian approximation for each slice for every distance

*rsquared* - quality of Gaussian approximatation (the r>0.98 is good). It is show, where pulse is gaussian

*center_x* and *center_y* - center of gaussian distribution (in pixels)
*1/width_x* and *1/width_y* - The back value to gauthe bigger values - the closer to waist

*slices_intensity* - total intensity for each slice

Coordinates on z-axis  - distance from original spike

**For 1D plots**:

*pulse_rsquared* - quality of Gaussian approximatation (the r>0.98 is good) for total pulse (calculated by total pulse intensity)

*pulse_center_x* and *pulse_center_y* - center of gaussian distribution (in pixels) for total pulse depending on distance

*pulse_width_x* and *pulse_width_y* + zoomed zone of minimum width- width of gaussian distribution (in pixels) for total pulse depending on distance

*Intensity per pixel* + zoomed zone of maximum intensity - the intensity of total pulse at 25%, 50% and 75% from gaussian height, devided by pixels count and by maximum intesity (for comfortble seching maximum position)

```python
#TODO: check limits and sizes
for key in collected_data:
    print key
    data = collected_data[key]
    res={}
    gaussian_options = ['height', 'rsquared', 'width_x', 'width_y','center_x', 'center_y','slices_intensity',
                        'pulse_rsquared','pulse_center_x','pulse_center_y','pulse_width_x', 'pulse_width_y',
                        'pulse_total_intensity_0_25','pulse_total_intensity_0_50','pulse_total_intensity_0_75',
                        'pulse_variance_x','pulse_variance_y','pulse_variance', 'pulse_max_intensity']
    
    for gaussian_param in gaussian_options:
        z=[]
        if data.values()[0]['misc']['gaussian_parameters/'+gaussian_param].ndim == 0:
            res[gaussian_param] = numpy.zeros(len(data.keys()))
        else:
            res[gaussian_param] = numpy.zeros((len(data.keys()),
                                len(data.values()[0]['misc']['gaussian_parameters/'+gaussian_param])))
            
        for ik, k in enumerate(data.iterkeys()):
            z.append(float(k))
            if  res[gaussian_param].ndim == 1: # for 1d arrays
                res[gaussian_param][ik]=data[k]['misc']['gaussian_parameters/'+gaussian_param]
            else:
                res[gaussian_param][ik,:]=data[k]['misc']['gaussian_parameters/'+gaussian_param]
    z=np.array(z)
    
    fig_size = (12,8)
    
    
    for ig, gaussian_param in enumerate(gaussian_options[:-6]):
        fig = plt.figure(figsize=fig_size);
        if res[gaussian_param].ndim == 2:  # for 2d arrays
            extents = [0, res[gaussian_param].shape[1], z.min(), z.max()]
            
            X,Y = np.meshgrid(np.arange(extents[1]),z)
            Xs,Ys=np.meshgrid(np.arange(extents[1]),np.linspace(extents[2],extents[3],200))
            
            res_resampled = plt.griddata(X.flatten(), Y.flatten(), res[gaussian_param].flatten(), Xs, Ys)
            
            
            if gaussian_param == 'rsquared':
                plt.pcolor(Xs, Ys, res_resampled, vmin=0.98);
                plt.title(gaussian_param);
            elif gaussian_param.startswith('width'):
                plt.pcolor(Xs, Ys, 1./res_resampled);
                plt.title('1/'+gaussian_param);
            else:
                plt.pcolor(Xs, Ys, res_resampled);
                plt.title(gaussian_param);
                
            plt.ylabel('z, m');
            plt.xlabel('Slice number');
            plt.axis('tight');
            plt.colorbar();
            
        elif res[gaussian_param].ndim == 1:  # for 1d arrays
            if ('pulse_total_intensity_' in gaussian_param) or ('pulse_width_' in gaussian_param) or ('pulse_center_' in gaussian_param):
                plt.close(fig)
                continue
                
            plt.plot(z, res[gaussian_param],'*');
            plt.grid(True);
            if 'pulse_rsquared' in gaussian_param:
                plt.ylabel('r^2');
            else:
                plt.ylabel('px');
            plt.title(gaussian_param);
        else:
            raise TypeError('Support only 1D and 2D arrays')
        plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)));
        plt.show();
        
    plt.figure(figsize=fig_size);
    plt.plot(z, res['pulse_center_x']*pixel_size_x,'*', label='pulse_center_x, $\mu$ m')
    plt.ylabel('Pulse center position, $\mu$ m')
    plt.xlabel('z, m')
    plt.grid(True)
    pylab.legend()
    plt.title('Pulse center, x')
    plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)))
    plt.show()
    
    plt.figure(figsize=fig_size);
    plt.plot(z, res['pulse_center_y']*pixel_size_y,'*', label='pulse_center_y, $\mu$ m');
    plt.ylabel('Pulse center position, $\mu$ m')
    plt.xlabel('z, m')
    plt.grid(True)
    pylab.legend()
    plt.title('Pulse center, y')
    plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)))
    plt.show()
    
    plt.figure(figsize=fig_size);
    min_x_position_index = np.argmin(res['pulse_width_x'])
    min_x_position = z[min_x_position_index]
    min_x_value = res['pulse_width_x'][min_x_position_index]*pixel_size_x
    
    plt.plot(z, res['pulse_width_x']*pixel_size_x,'*',
             label='pulse width, x axis, (minimum value {:0.2f} $\mu m$ at position {:0.2f} m)'.format(
                min_x_value, min_x_position)
             );
    plt.hold(True)
    plt.axvline(min_x_position, color='b')
 
    
    min_y_position_index = np.argmin(res['pulse_width_y'])
    min_y_position = z[min_y_position_index]
    min_y_value = res['pulse_width_y'][min_y_position_index]*pixel_size_y
    
    plt.plot(z, res['pulse_width_y']*pixel_size_y,'r*',
             label='pulse width, y axis, (minimum value {:0.2f} $\mu m$ at position {:0.2f} m)'.format(
            min_y_value, min_y_position));
    plt.axvline(min_y_position,color='r')
    
    plt.grid(True)
    plt.ylabel('Width, $\mu$ m')
    plt.xlabel('z, m')
    plt.grid(True)
    pylab.legend()
    plt.title('Pulse width')
    
    plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)))
    plt.show()
    
    plt.figure(figsize=fig_size);
    
    x_range=[-20,5] # show only this range
    
    z_mask = (x_range[0]<z)*(z<x_range[1])
    z_masked = z[z_mask]
    value_x = res['pulse_width_x'][z_mask]*pixel_size_x
    value_y = res['pulse_width_y'][z_mask]*pixel_size_y
        
    min_x_position_index = np.argmin(res['pulse_width_x'])
    min_x_position = z[min_x_position_index]
    min_x_value = res['pulse_width_x'][min_x_position_index]*pixel_size_x
    
    plt.plot(z_masked, value_x,'*',
            label='pulse width, x axis, (minimum value {:0.2f} $\mu m$ at position {:0.2f} m)'.format(
                min_x_value, min_x_position));
    plt.hold(True)
    plt.axvline(min_x_position, color='b')
    
    
    min_y_position_index = np.argmin(res['pulse_width_y'])
    min_y_position = z[min_y_position_index]
    min_y_value = res['pulse_width_y'][min_y_position_index]*pixel_size_y
    
    plt.plot(z_masked, value_y,'r*',
            label='pulse width, y axis, (minimum value {:0.2f} $\mu m$ at position {:0.2f} m)'.format(
            min_y_value, min_y_position));
    plt.axvline(min_y_position,color='r')
    
    plt.grid(True)   
    plt.ylabel('Width, , $\mu m$')
    plt.xlabel('z, m')
    pylab.legend()
    plt.title('Pulse width (zoomed in minimum width zone)')
    
    
    plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)))
    plt.show()
    
    plt.figure(figsize=fig_size);
    
    #area = (res['pulse_width_x']*res['pulse_width_y'])
    area = (res['pulse_width_x']*res['pulse_width_y']) #number of pixels
     
    J2eV = 6.24150934e18
    
#    data = res['pulse_total_intensity_0_25']/(res['pulse_total_intensity_0_25'].max()*area)
    data = res['pulse_total_intensity_0_25']/(area*16)*pixel_size_x*pixel_size_y*1e-6*dt*J2eV/phE # [mm^2]
    max_x_position_index = np.argmax(data)
    max_x_position = z[max_x_position_index]
    
    plt.plot(z, data ,'*', label=
             'Cut level: 25% (max value at position {:0.1f} m)'.format(max_x_position));
    plt.hold(True)
    plt.axvline(max_x_position, color='b')
    
    #data = res['pulse_total_intensity_0_50']/(res['pulse_total_intensity_0_50'].max()*area)
    data = res['pulse_total_intensity_0_50']/(area*4)*pixel_size_x*pixel_size_y*1e-6*dt*J2eV/phE # [mm^2]
    max_x_position_index = np.argmax(data)
    max_x_position = z[max_x_position_index]
    
    plt.plot(z, data ,'r*', label=
             'Cut level: 50% (max value at position {:0.1f} m)'.format(max_x_position));
    plt.axvline(max_x_position, color='r')
    
    #data = res['pulse_total_intensity_0_75']/(res['pulse_total_intensity_0_75'].max()*area)
    data = res['pulse_total_intensity_0_75']/area*pixel_size_x*pixel_size_y*1e-6*dt*J2eV/phE # [mm^2]
    max_x_position_index = np.argmax(data)
    max_x_position = z[max_x_position_index]
    
    plt.plot(z, data ,'g*', label=
             'Cut level: 75% (max value at position {:0.1f} m)'.format(max_x_position));
    plt.axvline(max_x_position, color='g')
    
    plt.grid(True)
    plt.xlabel('z, m')
    plt.ylabel('Number of photons')
    pylab.legend(loc=2)
    plt.title('Average number of photons per '+str(np.floor(pixel_size_x)) +'x'+
              str(np.floor(pixel_size_y))+'$\mu m^2$ pixel')
    
    plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)))
    plt.show()
    
    plt.figure(figsize=fig_size);
    area = (res['pulse_width_x']*res['pulse_width_y'])
    
    #data = res['pulse_total_intensity_0_25']/(res['pulse_total_intensity_0_25'].max()*area)
    data = res['pulse_total_intensity_0_25']/(area*16)*pixel_size_x*pixel_size_y*1e-6*dt*J2eV/phE # [mm^2]
    max_x_position_index = np.argmax(data)
    max_x_position = z[max_x_position_index]
    
    plt.plot(z[z_mask], data[z_mask] ,'*', label=
             'Cut level: 25% (max value at position {:0.1f} m)'.format(max_x_position));
    plt.hold(True)
    plt.axvline(max_x_position, color='b')
    
    #data = res['pulse_total_intensity_0_50']/(res['pulse_total_intensity_0_50'].max()*area)
    data = res['pulse_total_intensity_0_50']/(area*4)*pixel_size_x*pixel_size_y*1e-6*dt*J2eV/phE # [mm^2]
    max_x_position_index = np.argmax(data)
    max_x_position = z[max_x_position_index]
    
    plt.plot(z[z_mask], data[z_mask] ,'r*', label=
             'Cut level: 50% (max value at position {:0.1f} m)'.format(max_x_position));
    plt.axvline(max_x_position, color='r')
    
    #data = res['pulse_total_intensity_0_75']/(res['pulse_total_intensity_0_75'].max()*area)
    data = res['pulse_total_intensity_0_75']/area*pixel_size_x*pixel_size_y*1e-6*dt*J2eV/phE # [mm^2]
    max_x_position_index = np.argmax(data)
    max_x_position = z[max_x_position_index]
    
    plt.plot(z[z_mask], data[z_mask] ,'g*', label=
             'Cut level: 75% (max value at position {:0.1f} m)'.format(max_x_position));
    plt.axvline(max_x_position, color='g')
    
    plt.grid(True)
    plt.xlabel('z, m')
    plt.ylabel('Number of photons')
    #pylab.legend(loc=2)
    plt.title('Average number of photons per '+str(np.floor(pixel_size_x)) +'x'+
              str(np.floor(pixel_size_y))+'$\mu m^2$ pixel'+' (maximum intensity zone)')
    
    plt.savefig(os.path.join(root_directory,'{:0.4f}_{}.png'.format(key, gaussian_param)))
    plt.show()
    
    
```

**Gausiian beam quality - M^2 factor**

$ \sigma^2(z) = \sigma_0^2 + M^4 \left(\frac{\lambda}{\pi\sigma_0}\right)^2(z-z_0)^2 $  (https://en.wikipedia.org/wiki/Beam_parameter_product)

```python
import scipy.optimize as optimization
```

```python
def m2(z, s0,M,z0):
    """Returns a gaussian function with the given parameters"""
    energy = 5.0 # keV
    l = 12.398e-10/energy  # lambda in m
    return np.sqrt(np.abs(s0*s0+(M**4)*(l/(np.pi*s0))**2*(z-z0)**2))
```

Calculating beam parameters using Gaussian fitting

```python
# filter data (gaussian width)
xf=z
yf=res['pulse_width_x']*pixel_size_x

# indexes = np.isclose((xf-xf.min())%5,0)
# xf=xf[indexes]
# yf=yf[indexes]

fit_res = scipy.optimize.curve_fit(m2, xf,2*yf*1e-6)
#print res[0]
s0, M, z0 = fit_res[0]
print 'Waist size = ',s0*1e6 , ',mkm'
print 'M^2 = ',M**2
print 'Waist position = ',z0 , ',m'

x_sorted = np.sort(xf)
m = [m2(xx, s0,M,z0) for xx in x_sorted]
m=np.array(m)
plt.plot(xf,yf,'*')
plt.hold(True)
plt.plot(x_sorted,m*1e6/2, 'r', linewidth=3)
plt.grid(True)
```
